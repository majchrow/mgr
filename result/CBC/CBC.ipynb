# CBC algorythm for lattice points
[Constructing Randomly Shifted Lattice Rules in Weighted Sobolev Spaces(I. H. Sloan, F. Y. Kuo, and S. Joe)](https://doi.org/10.1137/S0036142901393942)

Naive implementation with time complexity O(n^2 * d_max)

```python
## w(i) = 0.9^(i)
## d_max(max dimension) = 100 | n - number of total grid points
## w1 <- n = 2003  | ~ 5  seconds
## w2 <- n = 8009  | ~ 92 seconds
## w3 <- n = 32003 | ~ 1507 seconds

w1 = pd.read_table("w1.txt", sep=" ")
w2 = pd.read_table("w2.txt", sep=" ")
w3 = pd.read_table("w3.txt", sep=" ");
```

```python
plt.figure(figsize=(20,10))
plt.plot(w1.d, w1["D(z)"], "x", label="n=2003")
plt.plot(w2.d, w2["D(z)"], "*", label="n=8009")
plt.plot(w3.d, w3["D(z)"], "+", label="n=32003")
plt.legend(loc="best")
plt.xlabel("Dimension")
plt.ylabel("Error")
plt.show()
```
